```python
asorted = BarcavsMadrid.sort_values('Team', ignore_index=True)
asorted.Team.value_counts()
f0, f1 = asorted.Team.value_counts()
fig, axs = plt.subplots(figsize = (12.5,10))

axs1 = plt.subplot2grid(shape = (3,3), loc = (0,0))
axs2 = plt.subplot2grid(shape = (3,3), loc = (0,1))
axs3 = plt.subplot2grid(shape = (3,3), loc = (0,2))
axs4 = plt.subplot2grid(shape = (3,3), loc = (1,0))
axs5 = plt.subplot2grid(shape = (3,3), loc = (1,1))
axs6 = plt.subplot2grid(shape = (3,3), loc = (1,2))
axs7 = plt.subplot2grid(shape = (3,3), loc = (2,0))
axs8 = plt.subplot2grid(shape = (3,3), loc = (2,1))
axs9 = plt.subplot2grid(shape = (3,3), loc = (2,2))


plt.tight_layout()

axs1.hist(BarcavsMadrid.iloc[0:f0,0], edgecolor = 'b', fc = 'none', label = 'FC Barcelona')
axs1.hist(BarcavsMadrid.iloc[f0:f0+f1,0], edgecolor = 'r', fc = 'none', label = 'Real Madrid')
axs1.set_xlabel('Win/Loss (a)')
axs1.set_ylabel('Win/Loss')
axs1.legend()

axs2.scatter(asorted.iloc[0:f1, 1], asorted.iloc[0:f1, 0], label = 'FC Barcelona')
axs2.scatter(asorted.iloc[f1:f0+f1, 1], asorted.iloc[f1:f0+f1, 0], label = 'Real Madrid')
axs2.set_xlabel('Goals Scored (b)')
axs2.set_ylabel('WIn/Loss')
axs2.legend()

axs3.scatter(asorted.iloc[0:f1, 2], asorted.iloc[0:f1, 0], label = 'FC Barcelona')
axs3.scatter(asorted.iloc[f1:f0+f1, 2], asorted.iloc[f1:f0+f1, 0], label = 'Real Madrid')
axs3.set_xlabel('Goals Against (c)')
axs3.set_ylabel('Win/Loss')
axs3.legend()

axs4.scatter(asorted.iloc[0:f1, 0], asorted.iloc[0:f1, 1], label = 'FC Barcelona')
axs4.scatter(asorted.iloc[f1:f0+f1, 0], asorted.iloc[f1:f0+f1, 1], label = 'Real Madrid')
axs4.set_xlabel('Win/Loss (d)')
axs4.set_ylabel('Goals Scored')
axs4.legend()

axs5.hist(BarcavsMadrid.iloc[0:f0,1], edgecolor = 'b', fc = 'none', label = 'FC Barcelona')
axs5.hist(BarcavsMadrid.iloc[f0:f0+f1,1], edgecolor = 'r', fc = 'none', label = 'Real Madrid')
axs5.set_xlabel('Goals Scored (e)')
axs5.set_ylabel('Goals Scored')
axs5.legend()

axs6.scatter(asorted.iloc[0:f1, 2], asorted.iloc[0:f1, 1], label = 'FC Barcelona')
axs6.scatter(asorted.iloc[f1:f0+f1, 2], asorted.iloc[f1:f0+f1, 1], label = 'Real Madrid')
axs6.set_xlabel('Goals Against (f)')
axs6.set_ylabel('Goals Scored')
axs6.legend()

axs7.scatter(asorted.iloc[0:f1, 0], asorted.iloc[0:f1, 2], label = 'FC Barcelona')
axs7.scatter(asorted.iloc[f1:f0+f1, 0], asorted.iloc[f1:f0+f1, 2], label = 'Real Madrid')
axs7.set_xlabel('Win/Loss (g)')
axs7.set_ylabel('Goals Against')
axs7.legend()

axs8.scatter(asorted.iloc[0:f1, 1], asorted.iloc[0:f1, 2], label = 'FC Barcelona')
axs8.scatter(asorted.iloc[f1:f0+f1, 1], asorted.iloc[f1:f0+f1, 2], label = 'Real Madrid')
axs8.set_xlabel('Goals Scored (h)')
axs8.set_ylabel('Goals Against')
axs8.legend()

axs9.hist(BarcavsMadrid.iloc[0:f0,2], edgecolor = 'b', fc = 'none', label = 'FC Barcelona')
axs9.hist(BarcavsMadrid.iloc[f0:f0+f1,2], edgecolor = 'r', fc = 'none', label = 'Real Madrid')
axs9.set_xlabel('Goals Against( i)')
axs9.set_ylabel('Goals Against')
axs9.legend()
```
